```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
x = np.random.rand(100) * 10

noise = np.random.normal(0,1,size=len(x))
y = 2*x + 3 +noise

plt.figure(figsize=(8,6))
plt.scatter(x,y,color = 'blue')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot')
plt.legend(['Scatter'])
plt.grid(True)
plt.show()
```
