```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = [1,2,2,3,3,4,4,5,5,5,6]
y = range(1,len(x)+1)

plt.plot(x, y, linestyle='--', marker='o', color='k')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Simple Line Plot')
plt.legend(['Data Line'])
plt.grid(True)
plt.show()
```
