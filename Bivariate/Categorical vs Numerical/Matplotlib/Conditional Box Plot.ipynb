```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame(
    {'Team A': np.random.randint(20,50,30),
    'Team B': np.random.randint(40,70,30)}
)
df.head()
```

```python
plt.boxplot(df,labels = df.columns) 
plt.xlabel('Team')
plt.ylabel('Score')
plt.title('Boxplots by groups')
plt.grid(True)
plt.show()
```

```python
data = {
    'Category': ['A', 'A', 'B', 'B', 'C', 'C', 'A', 'B', 'C', 'A', 'B', 'C'],
    'Value': [10, 15, 10, 20, 30, 35, 20, 25, 30, 15, 20, 25]
}

df = pd.DataFrame(data)

# Prepare data for plt.boxplot
grouped = df.groupby('Category')['Value'].apply(list)
data_to_plot = [group for group in grouped]

# Create a box plot
plt.boxplot(data_to_plot, labels=grouped.index)

plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Box Plot by Category')
plt.show()
```
