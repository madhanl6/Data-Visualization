```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
categories = ['Category 1', 'Category 2', 'Category 3']
values_group1 = [10, 20, 30]
values_group2 = [15, 25, 5]

# Set the position of bars on x-axis
x = np.arange(len(categories))

# Bar width
width = 0.6

# Plot the first group of bars
plt.bar(x, values_group1, width, label='Group 1', color='blue')

# Plot the second group of bars, stacked on top of the first
plt.bar(x, values_group2, width, bottom=values_group1, label='Group 2', color='orange')

plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Stacked Bar Chart of Two Groups')
plt.xticks(x, categories)
plt.legend()
plt.show()
```
