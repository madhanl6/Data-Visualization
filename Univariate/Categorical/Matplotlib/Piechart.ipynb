```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
values = [577,314]
index = ['Male','Female']
columns = ['Passengers']

df = pd.DataFrame(values,index=index,columns=columns)
df
```

```python
plt.pie(x=df['Passengers'],
        labels =df.index,
        autopct='%0.2f%%',
        explode=[0.1,0],
        colors=['lightblue', 'lightcoral'])
plt.title("Distribution of Passangers by Gender")
plt.show()
```

```python
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Basic Pie Chart')
plt.show()
```
