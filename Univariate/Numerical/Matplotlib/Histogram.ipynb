```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 2, 3, 3, 4, 4, 5, 5, 5, 6]

plt.hist(x, bins=5, density=False, color='green', edgecolor='black', alpha=0.7)
plt.xlabel('X')
plt.ylabel('Frequency') 
plt.xlim(0, 8)
plt.ylim(0, 10)
plt.title('Histogram Plot')
plt.legend(['hist'])
plt.grid(True)
plt.show()
```

```python
data = np.random.randn(1000)

plt.hist(data, bins=30, edgecolor='black')

plt.title('Histogram Example')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```
