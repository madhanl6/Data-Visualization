```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

rcParams (short for runtime configuration parameters) is a way to customize and control the appearance of plots and figures. You can use rcParams to set global properties for all plots, such as figure size, line width, font size, and colors. This is useful for ensuring consistency across different plots and for making global adjustments to the aesthetics of your visualizations.

```python
# Set custom rcParams for a more appealing plot
plt.rcParams['figure.figsize'] = (8, 6)      # Set the figure size
plt.rcParams['axes.labelsize'] = 12          # Set font size for axis labels
plt.rcParams['axes.titlesize'] = 16          # Set font size for plot title
plt.rcParams['lines.linewidth'] = 2.5        # Set line width
plt.rcParams['lines.marker'] = 'o'           # Set data point marker
plt.rcParams['lines.markersize'] = 8         # Set marker size
plt.rcParams['axes.grid'] = True             # Display grid lines

# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 8, 6, 4, 2]

# Create a plot
plt.plot(x, y, label='Sample Data', color='b', linestyle='-', markeredgecolor='r')

# Add title and labels
plt.title('Customized Matplotlib Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

rcParams und Subplots

```python
# Set custom rcParams for consistent styling across all plots
plt.rcParams['figure.figsize'] = (12, 8)  # Set the figure size
plt.rcParams['axes.labelsize'] = 12       # Set font size for axis labels
plt.rcParams['axes.titlesize'] = 14       # Set font size for plot titles
plt.rcParams['lines.linewidth'] = 2.5     # Set line width
plt.rcParams['lines.marker'] = 'o'        # Set data point marker
plt.rcParams['lines.markersize'] = 8      # Set marker size
plt.rcParams['axes.grid'] = True          # Display grid lines

# Generate some synthetic data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(-x)

# Create a figure and subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# Plot data on the first subplot
axs[0, 0].plot(x, y1, color='b', label='sin(x)')
axs[0, 0].set_title('Sine Function')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('sin(x)')
axs[0, 0].legend()
axs[0, 0].grid(True)

# Plot data on the second subplot
axs[0, 1].plot(x, y2, color='r', linestyle='--', label='cos(x)')
axs[0, 1].set_title('Cosine Function')
axs[0, 1].set_xlabel('x')
axs[0, 1].set_ylabel('cos(x)')
axs[0, 1].legend()
axs[0, 1].grid(True)

# Plot data on the third subplot
axs[1, 0].plot(x, y3, color='g', linestyle='-.', label='tan(x)')
axs[1, 0].set_title('Tangent Function')
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel('tan(x)')
axs[1, 0].legend()
axs[1, 0].grid(True)
axs[1, 0].set_ylim(-10, 10)  # Limit y-axis to avoid extreme values

# Plot data on the fourth subplot
axs[1, 1].plot(x, y4, color='m', linestyle=':', marker='s', label='exp(-x)')
axs[1, 1].set_title('Exponential Decay')
axs[1, 1].set_xlabel('x')
axs[1, 1].set_ylabel('exp(-x)')
axs[1, 1].legend()
axs[1, 1].grid(True)

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```
