```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

plt.subplots() is a function that creates a figure and a set of subplots (axes) in a single call. It returns a tuple containing the figure and the array of axes objects. This method is useful when you want to create multiple subplots in a grid layout.

```python
# Create a 2x2 grid of subplots with a specific figure size
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))

# Data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = x**2

# Plotting on each subplot
axs[0, 0].plot(x, y1, 'r-')  # Plot sine wave
axs[0, 0].set_title('Sine')   # Title for the top-left subplot

axs[0, 1].plot(x, y2, 'g-')  # Plot cosine wave
axs[0, 1].set_title('Cosine') # Title for the top-right subplot

axs[1, 0].plot(x, y3, 'b-')  # Plot tangent wave
axs[1, 0].set_title('Tangent') # Title for the bottom-left subplot

axs[1, 1].plot(x, y4, 'm-')  # Plot quadratic function
axs[1, 1].set_title('Quadratic') # Title for the bottom-right subplot

# Add a global title for the entire figure
fig.suptitle('Various Trigonometric and Polynomial Functions', fontsize=16)

# Adjust layout to prevent overlap and ensure titles fit
plt.tight_layout(rect=[0, 0, 1, 0.96])

# Display the plot
plt.show()
```
