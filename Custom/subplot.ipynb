```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

subplot is a function that allows you to create multiple plots within a single figure, organizing them in a grid. Itâ€™s a convenient way to compare different plots side-by-side or in a matrix layout.

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = x

# Create a new figure with a specified size
plt.figure(figsize=(14, 4))  # Adjusted figure size to accommodate 4 subplots in a row

# First subplot (1st row, 1st column)
plt.subplot(1, 4, 1)  # 1 row, 4 columns, position 1
plt.plot(x, y1, 'r')  # Plot sine wave in red
plt.title('Sine Wave')  # Title for the first subplot

# Second subplot (1st row, 2nd column)
plt.subplot(1, 4, 2)  # 1 row, 4 columns, position 2
plt.plot(x, y2, 'b')  # Plot cosine wave in blue
plt.title('Cosine Wave')  # Title for the second subplot

# Third subplot (1st row, 3rd column)
plt.subplot(1, 4, 3)  # 1 row, 4 columns, position 3
plt.plot(x, y3, 'g')  # Plot tangent wave in green
plt.title('Tangent Wave')  # Title for the third subplot

# Fourth subplot (1st row, 4th column)
plt.subplot(1, 4, 4)  # 1 row, 4 columns, position 4
plt.plot(x, y4, 'c')  # Plot linear function in cyan
plt.title('Linear')  # Title for the fourth subplot

# Adjust layout to prevent overlap between subplots
plt.tight_layout()

# Display the figure with all subplots
plt.show()
```
